# Determine if Brackets are Balanced

```python
def is_match(p1, p2):
    if p1 == "(" and p2 == ")":
        return True
    elif p1 == "{" and p2 == "}":
        return True
    elif p1 == "[" and p2 == "]":
        return True
    else:
        return False
```

```python
def is_paren_balanced(paren_string):
    """
    This function takes a string of parentheses as input and returns True if the parentheses are balanced, otherwise False.
    """
    s = []
    is_balanced = True
    index = 0

    while index < len(paren_string) and is_balanced:
        curr_elem = paren_string[index]
        if curr_elem in "({[":
            s.append(curr_elem)
        else:
            # if stack is empty, then it is not balanced as the starting bracket is not present
            if len(s) == 0:
                is_balanced = False
                break
            else:
                # if stack is not empty, then pop the top element and check if it is a match with the current element
                top_elem = s.pop()
                if not is_match(top_elem, curr_elem):
                    is_balanced = False
                    break
        index += 1

    if len(s) == 0 and is_balanced:
        return True
    else:
        return False
```

```python
test_string = "{[({})]}"
print(is_paren_balanced(test_string))

test_string = "{[}"
print(is_paren_balanced(test_string))

test_string = "}]"
print(is_paren_balanced(test_string))
```

# Reverse String (FILO)

```python
def reverse_string(input_str):
    """
    This function takes a string as input and returns the reverse of the string using a stack.
    """
    s = list()
    for elem in input_str:
        s.append(elem)

    reverse_str = ""
    while len(s) > 0:
        reverse_str += s.pop()

    return reverse_str
```

```python
test_str = "Hello"
print(reverse_string(test_str))
```

# Exercise: Convert Decimal Integer to Binary

```python
def convert_int_to_bin(input_num: int):
    if input_num ==0:
        return 0

    s = list()
    while input_num > 0:
        s.append(input_num % 2)
        input_num = input_num // 2

    rev = ""
    while len(s) > 0:
        rev += str(s.pop())

    return rev
```

```python
convert_int_to_bin(242)
```
